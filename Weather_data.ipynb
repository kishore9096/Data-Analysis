# Question and Answers 

```python
# Lets asks some questions about the data and try to answer them.
```

```python
# 1. Find the unique values in "Wind speed"?

df["Wind Speed_km/h"].unique()
```

```python
# 2. Find the number of times when the weather is eactly clear.

df[df["Weather"] == "Clear"]["Weather"].value_counts()
```

```python
# By using groupby

df.groupby("Weather").get_group("Clear").count()
```

```python
# 3. Find the number of times when the wind speed was exactly 4 kmph.

df[df["Wind Speed_km/h"] == 4].count()
```

```python
# 4. Rename the column name "Weather" to "Weather condition"

df.rename(columns = {"Weather" : "Weather conditions"})
```

```python
# Rename more than one column name

df.rename(columns = {"Weather" : "Weather conditions", "Temp_C" : "Temperature"},inplace = True)
```

```python
df.columns
```

```python
# 6. What is the mean of "Visibility"?

df.Visibility_km.mean()
```

```python
# 7. What is the sd of "Pressure" in this data?

df.Press_kPa.std()
```

```python
# 8. What is the variance of "Relative Humidity" in this data?

df["Rel Hum_%"].var()
```

```python
# 9. Find all instances where Snow was recorded.

df.tail(2)
```

```python
df[df["Weather conditions"] == "Snow"].count()  # This will show the rows where the column contains only snow, but we also have snow fog or snow mist columns. 
```

```python
# To solve the above problem we use string.contains

df[df["Weather conditions"].str.contains("Snow")].count()
```

```python
df[df["Weather conditions"] == "Snow"]["Temperature"].max()   # Some times snow starts from this temp.

df.columns
```

```python
# 10. Find all instances when "Wind speed is above 24" and "Visibility is 25".

df[(df["Wind Speed_km/h"] > 24) & (df["Visibility_km"] == 25)]
```

```python
# What is the Mean value of each column against each "Weather condition"? 

df["Weather conditions"].unique()
```

```python
df.groupby("Weather conditions").mean()
```

```python
# 12. What is the minimum and Maximum of each column against each weather condition.

df.groupby("Weather conditions").min()
```

```python
df.groupby("Weather conditions").max()
```

```python
# 15. Find all instances .A. Weather is clear and Relative Humidity is greater than 50 or Visibility is above 40.

df[(df["Weather conditions"] == "Clear") & (df["Rel Hum_%"] >= 50) | (df["Visibility_km"] > 40)]
```
